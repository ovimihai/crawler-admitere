```
save_path = 'raw/'
year = '2019'
```

```
headers = {'User-Agent': 'Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)'}
```

Descarca judete

```
url = 'http://static.admitere.edu.ro/{year}/repartizare/index.html'
r = requests.get(url.format(year=year), allow_redirects=True, headers=headers)
main = r.content

main = str(main).replace("\\r\\n", '').replace("\\n", '')

judete = re.findall(r'<div class="card"><a class="card-body" href="([^"]*)">([^"]*)<\/a><\/div>', main)
```

```
files = ['specialization', 'candidate', 'highschool', 'school', 'empty-seats']
# Downloading Raw data
for j in judete:
  for f in files:
    out_file = "{f}-{year}-{judet}.raw.json".format(f=f, year=year, judet=j[0])
    print("Downloading {file}".format(file=out_file))
    url = 'http://static.admitere.edu.ro/{year}/repartizare/{judet}/data/{f}.json'
    r = requests.get(url.format(f=f, year=year, judet=j[0]), allow_redirects=True, headers=headers)
    candidates_str = r.content
    print(len(r.content))
    open(save_path + out_file, 'wb').write(r.content)
```
